```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
plt.style.use('seaborn-colorblind')
```

```python
movies_data = pd.read_csv('movies.csv')
movies_data.head()
```

```python
ratings_data = pd.read_csv('ratings.csv')
ratings_data.head()
```

```python
avg_rate = ratings_data.groupby(['movieId']).mean()
avg_rate = avg_rate.drop(['userId', 'timestamp'], axis = 1)
avg_rate = avg_rate.rename(columns = {"rating" : "avg_rating"})
avg_rate.head()
```

```python
vote_count = ratings_data.groupby(['movieId']).count()
vote_count = vote_count.drop(['userId', 'timestamp'], axis = 1)
vote_count = vote_count.rename(columns = {"rating" : "No.of.Votes"})
vote_count.head()
```

```python
movies_data.merge(vote_count, how = 'outer')
```

```python
movies_data['VoteCount'] = pd.Series(vote_count['No.of.Votes'], index = movies_data.index)
movies_data['avg_rating'] = pd.Series(avg_rate['avg_rating'], index = movies_data.index)
movies_data.head()
```
